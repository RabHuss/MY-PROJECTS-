```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import geopandas as gpd
```

```python
df = pd.read_excel('DELVe.xlsx')

print(df.head())
```

```python

missing_data = df.isnull().sum()
print(missing_data)
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip().str.replace('_', ' ')
```

```python

disability_distribution = df['disability id'].value_counts()
print(disability_distribution)


plt.figure(figsize=(10,6))
sns.countplot(x='disability id', data=df, palette='viridis')
plt.title('Distribution of Students with Disabilities')
plt.xlabel('Disability ID')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

```python


class_distribution = df.groupby('class id')['disability id'].count()
print(class_distribution)


plt.figure(figsize=(10,6))
sns.countplot(x='class id', data=df, hue='disability id', palette='viridis')
plt.title('Classroom Distribution of Students with Disabilities')
plt.xlabel('Class ID')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns = df.columns.str.strip()  # Removes any extra spaces
```

```python


lgea_distribution = df.groupby('lgea')['disability id'].count()
print(lgea_distribution)


plt.figure(figsize=(12,6))
sns.countplot(x='lgea', data=df, palette='viridis', order=df['lgea'].value_counts().index)
plt.title('Geographical Distribution of Students with Disabilities (by LGEA)')
plt.xlabel('LGEA')
plt.ylabel('Number of Students')
plt.xticks(rotation=90)
plt.show()
```
